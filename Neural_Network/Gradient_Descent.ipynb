```python
import torch
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2

# Gradient descent with momentum
def gradient_descent(start_x, learning_rate, epochs):
    x = torch.tensor([start_x], requires_grad=True)
    # Set the learning rate
    x_list = [x.item()]

    for _ in range(epochs):
        y = f(x)
        y.backward()
        with torch.no_grad():
            x -= learning_rate * x.grad
            x.grad.zero_()
            x_list.append(x.item())

    return x_list

# Parameters
start_x = 3.0
learning_rate = 0.1

epochs = 100

# Run gradient descent with momentum
x_list = gradient_descent(start_x, learning_rate, epochs)

# Plot the results
plt.figure(figsize=[10,3])

# Cost function plot
plt.subplot(1, 2, 1)
x = torch.linspace(-3, 3, 100)
plt.plot(x.numpy(), f(x).numpy())
plt.scatter(x_list, f(torch.tensor(x_list)).numpy(), color='red')
plt.title('Cost function')

# Convergence plot
plt.subplot(1, 2, 2)
plt.plot(f(torch.tensor(x_list)).numpy())
plt.title('Convergence')

plt.show()
```

# Momentum

```python
import torch
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2

# Gradient descent with momentum
def gradient_descent_with_momentum(start_x, learning_rate, gamma, epochs):
    x = torch.tensor([start_x], requires_grad=True)
    v = torch.tensor([0.0])
    x_list = [x.item()]

    for _ in range(epochs):
        y = f(x)
        y.backward()
        with torch.no_grad():
            v = gamma * v + learning_rate * x.grad
            x -= v
            x.grad.zero_()
            x_list.append(x.item())

    return x_list

# Parameters
start_x = 3.0
learning_rate = 0.1
gamma = 0.9
epochs = 100

# Run gradient descent with momentum
x_list = gradient_descent_with_momentum(start_x, learning_rate, gamma, epochs)

# Plot the results
plt.figure(figsize=[10,3])

# Cost function plot
plt.subplot(1, 2, 1)
x = torch.linspace(-3, 3, 100)
plt.plot(x.numpy(), f(x).numpy())
plt.scatter(x_list, f(torch.tensor(x_list)).numpy(), color='red')
plt.title('Cost function')

# Convergence plot
plt.subplot(1, 2, 2)
plt.plot(f(torch.tensor(x_list)).numpy())
plt.title('Convergence')

plt.show()
```
