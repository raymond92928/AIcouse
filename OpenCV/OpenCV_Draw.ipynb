```python
import cv2
import numpy as np
```

# cv2.line(image, start_point, end_point, color, thickness)

```python
gc = np.zeros((512,512,3), dtype = np.uint8)
gc.fill(255)
cv2.line(gc, (0, 0), (400, 300), (255, 0, 0), 5)

cv2.imshow("draw", gc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.rectangle(image, start_point, end_point, color, thickness)

```python
gc = np.zeros((512,512,3), dtype = np.uint8)
gc.fill(255)
cv2.rectangle(gc, (30, 50), (200, 280), (0, 0, 255), 5)
cv2.rectangle(gc, (100, 250), (200, 280), (0, 255, 255), -1)

cv2.imshow("draw", gc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.circle(image, center_coordinates, radius, color, thickness)

```python
gc = np.zeros((512,512,3), dtype = np.uint8)
gc.fill(255)
cv2.circle(gc, (100, 90), 80, (0, 255, 255), -1)
cv2.circle(gc, (200, 90), 50, (0, 0, 255), 5) 

cv2.imshow("draw", gc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
 


```python
gc = np.zeros((512,512,3), dtype = np.uint8)
gc.fill(255)
# font = cv2.FONT_HERSHEY_SIMPLEX
font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.putText(gc, "Raymond", (10,300), font, 2, (0,0,0), 2, cv2.LINE_AA) #LINE_AA can let text smooth
cv2.imshow("draw", gc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
