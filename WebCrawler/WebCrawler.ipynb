```python
import time
from selenium import webdriver #open googledriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
```

Load the chromedriver and open it

```python
PATH = "chromedriver"
driver = webdriver.Chrome(PATH)
```

Load the link

```python
driver.get("http://www.macaodaily.com/")
```

```python
button = driver.find_element_by_xpath('//*[@id="Layer1"]/table[1]/tbody/tr[2]/td/a[3]')
button.click()
```

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
titles = driver.find_elements_by_id("all_article_list")
for title in titles:
    print(title.text)
```

find the elements until more than 10 seconds

```python
titles = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.ID, "all_article_list"))
)
#titles = driver.find_elements_by_id("all_article_list")
for title in titles:
    print(title.text)
```

```python
driver.quit()
```
