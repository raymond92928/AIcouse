```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from mpl_toolkits import mplot3d
from sklearn.preprocessing import StandardScaler
```

```python
X_train = np.array([
    [158, 1],
    [170, 1],
    [183, 1],
    [191, 1],
    [155, 0],
    [163, 0],
    [180, 0],
    [158, 0],
    [170, 0]
])
y_train = [64, 86, 84, 80, 49, 59, 67, 54, 67]

X_test = np.array([
    [168, 1],
    [180, 1],
    [160, 0],
    [169, 0]
])
y_test = [65, 96, 52, 67]

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X_train[:,0], X_train[:,1], y_train, color='black')
ax.set_xlabel('Height in cm')
ax.set_ylabel('Gender(0 is male, 1 is female)')
ax.set_zlabel('Weight in kg')

ax.view_init(-160, 60)
```

```python
X_train = np.array([
    [158, 1],
    [170, 1],
    [183, 1],
    [191, 1],
    [155, 0],
    [163, 0],
    [180, 0],
    [158, 0],
    [170, 0]
])
y_train = [64, 86, 84, 80, 49, 59, 67, 54, 67]

X_test = np.array([
    [168, 1],
    [180, 1],
    [160, 0],
    [169, 0]
])
y_test = [65, 96, 52, 67]

K = 3
clf = KNeighborsRegressor(n_neighbors=K)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

print('Actual weights: %s' % y_test)
print('Predicted weights: %s' % predictions)
print('Mean absolute error: %s' % mean_absolute_error(y_test, predictions))
print('Mean squared error: %s' % mean_squared_error(y_test, predictions))
```

# Absolute Values of the errors(MAE)
$MAE = \dfrac{1}{n}\sum_{i=0}^{n-1}|y_i - \hat{y}_i|$ where $y_i$ is response variable and $\hat{y}_i$ is explanatory variable.

# Mean Squared Deviation(MSD)
$MSD = \dfrac{1}{n}\sum_{i=0}^{n-1}(y_i - \hat{y}_i)^2$ where $y_i$ is response variable and $\hat{y}_i$ is explanatory variable.
